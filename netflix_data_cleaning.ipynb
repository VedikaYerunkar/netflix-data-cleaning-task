```python
import pandas as pd

# Load the Netflix dataset (make sure the file name matches exactly)
df = pd.read_csv('netflix_titles.csv')

# Show first 5 rows
df.head()
```

```python
# Shape of dataset
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")

# Column names
print(df.columns)

# Missing values
print(df.isnull().sum())

# Check for duplicate rows
print("Duplicate Rows:", df.duplicated().sum())

# Data types
print(df.dtypes)
```

```python
# Check percentage of missing data
missing_percent = (df.isnull().sum() / len(df)) * 100
print(missing_percent)
```

```python
# Fill missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Not listed', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna('Not Rated', inplace=True)

# Drop rows with null in date_added and duration (very few)
df.dropna(subset=['date_added', 'duration'], inplace=True)
```

```python
# Convert to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
# Final check
print(df.isnull().sum())
print(df.dtypes)
print(df.head())
```

```python
# Save cleaned dataset
df.to_csv('netflix_cleaned.csv', index=False)
```
